# Using Nonderterministic Finite Automata To Generate Chord Progressions

The set of popular chrod progressions is a regular language.

This suggests that we can generate reasonable nice sounding music with something as simple as a finite automaton.

```python
# Using the chart above we can write the transition function as a dict
t = {
    1: {1,2,3,4,5,6},
    6: {4,5,1,2},
    4: {1,5,6,3},
    5: {1,4,6},
    3: {6,1,4,2},
    2: {5,4,3}  
}
```

```python
# We can visualise the automaton this defines using the python library graphviz

from graphviz import Digraph
from IPython import display

g = Digraph("Chord NFA")
for v in t:
    g.node(str(v))
    
for u in t:
    for v in t[u]:
        g.edge(str(u), str(v))

output_path = "./graph.png"
g.render(filename=output_path, format='png')
display.Image(filename=output_path)
```
